```python
--MARKALARIN MUSTERİ SAYILARINI VERİNİZ
SELECT  BRAND , COUNT(BRAND) MUSTERİ_SAYISI
FROM WEBOFFERS
GROUP BY BRAND
ORDER BY COUNT(BRAND) DESC
```

```python
-- HANGİ SEHİRDE NE KADAR ARAC BİLGİSİ VAR
SELECT CITY , COUNT(CITY) COUNT_
FROM CITY C
JOIN WEBOFFERS  W ON (W.CITYID = C.ID)
GROUP BY CITY
```

```python
-- İstanbul’da Volkswagen Passat marka bir araç arıyoruz.
--Kriterlerimiz şu şekilde.
--Kimden:Sahibinden
--Model:2014-2018 arası
--Vites:Otomatik
--Yakıt:Dizel
--Sıralama:Kilometre ve fiyata göre

SELECT * 
FROM WEBOFFERS W 
JOIN CITY C ON (W.CITYID = C.ID)
WHERE FROMWHO = 'SAHİBİNDEN' 
AND YEAR_ BETWEEN '2014'  AND  '2018'
AND SHIFTTYPE = 'OTOMATİK VİTES' 
AND  MODEL = 'PASSAT'
AND CITY = 'İSTANBUL'
ORDER BY KM , PRICE DESC
```

```python
--BMW marka araçları Ankara, İstanbul ve İzmir’de listelettiren sorguyu yazınız. Burada sorguya gönderilen şehir isimleri teker teker değil
--virgül ile birleştirilerek gönderilecek. ‘Ankara,İstanbul,İzmir’ şeklinde.

SELECT * 
FROM WEBOFFERS W 
JOIN CITY C ON (W.CITYID = C.ID)
WHERE CITY IN ('ANKARA' , 'İSTANBUL' , 'İZMİR') AND BRAND = 'BMW'
```

```python
--Araç markalarına göre her markadan ne kadar araç olduğu ve yüzdesel olarak toplamın ne kadarına tekabül ettiği bilgisini
--şekildeki gibi getiren sorguyu yazınız

select BRAND,Count(BRAND) musteri_sayi, count(*) * 100.0 / sum(count(*)) over() yuzdesi
from WEBOFFERS
group by BRAND
```
